```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 分析結果のsummaryから係数と列名を抽出
coef = result.params.values
column_names = result.params.index

# 'const' 列を除外
column_names = column_names.drop('const')

# 説明変数のデータフレームを作成
X = df[column_names]

# VIFを計算
vif_data = pd.DataFrame()
vif_data["feature"] = column_names
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(column_names))]

print(vif_data)
```
